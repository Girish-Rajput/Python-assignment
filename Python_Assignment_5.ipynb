```python
# Q1. Create a 2D NumPy array and calculate the sum of each row
import numpy as np
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(array_2d, axis=1)
print("Q1 - Row-wise sum:\n", row_sums)

# Q2. Find the mean of a specific column in a DataFrame
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30], 'B': [15, 25, 35]})
mean_b = df['B'].mean()
print("\nQ2 - Mean of column B:", mean_b)

# Q3. Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.figure()
plt.scatter(x, y)
plt.title("Q3 - Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Q4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
data = pd.DataFrame(np.random.rand(5, 3), columns=['A', 'B', 'C'])
correlation = data.corr()
plt.figure()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Q4 - Correlation Heatmap")
plt.show()

# Q5. Generate a bar plot using Plotly
import plotly.express as px
bar_data = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [10, 20, 15]})
fig = px.bar(bar_data, x='Category', y='Values', title='Q5 - Plotly Bar Plot')
fig.show()

# Q6. Create a DataFrame and add a new column based on an existing column
df = pd.DataFrame({'Value': [5, 10, 15]})
df['Double'] = df['Value'] * 2
print("\nQ6 - DataFrame with new column:\n", df)

# Q7. Perform element-wise multiplication of two NumPy arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 * arr2
print("\nQ7 - Element-wise multiplication:\n", result)

# Q8. Create a line plot with multiple lines using Matplotlib
plt.figure()
plt.plot([1, 2, 3], [1, 4, 9], label='Line 1')
plt.plot([1, 2, 3], [2, 5, 10], label='Line 2')
plt.title("Q8 - Multiple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

# Q9. Generate a DataFrame and filter rows where a column value is greater than a threshold
df = pd.DataFrame({'Score': [60, 75, 50, 90]})
filtered_df = df[df['Score'] > 70]
print("\nQ9 - Filtered DataFrame:\n", filtered_df)

# Q10. Create a histogram using Seaborn
plt.figure()
sns.histplot(df['Score'], bins=5, kde=True)
plt.title("Q10 - Histogram of Scores")
plt.show()

# Q11. Perform matrix multiplication using NumPy
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
matmul_result = np.dot(A, B)
print("\nQ11 - Matrix Multiplication:\n", matmul_result)

# Q12. Load a CSV file and display the first 5 rows
# Sample data for demonstration; in Colab use: pd.read_csv('filename.csv')
sample_df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [85, 90, 95]})
print("\nQ12 - First 5 rows of DataFrame:\n", sample_df.head())

# Q13. Create a 3D scatter plot using Plotly
import plotly.graph_objects as go
fig = go.Figure(data=[go.Scatter3d(
    x=[1, 2, 3], y=[4, 5, 6], z=[7, 8, 9],
    mode='markers',
    marker=dict(size=8, color='blue')
)])
fig.update_layout(title='Q13 - 3D Scatter Plot')
fig.show()
```
