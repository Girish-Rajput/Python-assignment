# RESTful API & Flask Assignment




### Question 1
What is a RESTful API?

### Answer
RESTful API is an architectural style for building web services that use HTTP methods to operate on resources. It follows principles like statelessness, resource-based URLs, and client-server architecture.

### Question 2
What is Flask, and why is it popular for building APIs?

### Answer
Flask is a lightweight Python web framework used for building applications and APIs. It is popular because it is simple, flexible, minimalistic, and supports many extensions.

### Question 3
What are HTTP methods used in RESTful APIs?

### Answer
Common HTTP methods: GET (fetch), POST (create), PUT (update), PATCH (partial update), DELETE (remove).

### Question 4
What is the purpose of the @app.route() decorator in Flask?

### Answer
It maps a URL path to a function. When the path is accessed, the function runs.

### Question 5
What is the role of Flask-SQLAlchemy?

### Answer
It is an ORM extension that simplifies interaction with databases using Python classes instead of SQL queries.

### Question 6
How do you create a basic Flask application?

### Answer
```python
from flask import Flask
app = Flask(__name__)
@app.route('/')
def home():
    return 'Hello Flask'
if __name__ == '__main__':
    app.run(debug=True)
```

### Question 7
How do you return JSON responses in Flask?

### Answer
```python
from flask import jsonify
@app.route('/json')
def send_json():
    return jsonify({'message': 'Success'})
```

### Question 8
How do you handle POST requests in Flask?

### Answer
```python
from flask import request, jsonify
@app.route('/submit', methods=['POST'])
def submit():
    data = request.json
    return jsonify(data)
```

### Question 9
How do you handle errors in Flask (e.g., 404)?

### Answer
```python
@app.errorhandler(404)
def not_found(e):
    return jsonify({'error': 'Not found'}), 404
```

### Question 10
How do you structure a Flask app using Blueprints?

### Answer
```python
from flask import Blueprint
users = Blueprint('users', __name__)
@users.route('/users')
def get_users():
    return 'Users List'
```
