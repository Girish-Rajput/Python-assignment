# Q1: What is the difference between interpreted and compiled languages?
"""
Compiled languages translate code into machine code before execution (e.g., C, C++), resulting in faster performance.
Interpreted languages execute code line by line using an interpreter (e.g., Python), making development faster and debugging easier.
"""

# Q2: What is exception handling in Python?
"""
Exception handling is the process of managing errors that occur during program execution.
Python provides try, except, finally, and else blocks to catch and handle exceptions gracefully without crashing the program.
"""

# Q3: What is the purpose of the finally block in exception handling?
"""
The finally block is used to execute cleanup code, such as closing files or releasing resources, regardless of whether an exception occurred.
"""

# Q4: What is logging in Python?
"""
Logging is a module used to track events during the execution of a program.
It helps in debugging and monitoring by storing information, warnings, errors, and debugging messages into files or the console.
"""

# Q5: What is the significance of the __del__ method in Python?
"""
The __del__ method is a destructor method that is called when an object is about to be destroyed.
It is used for cleanup activities like closing files or releasing resources.
"""

# Q6: What is the difference between import and from ... import in Python?
"""
- `import module`: Imports the whole module. You access functions with module.function().
- `from module import function`: Imports specific function or class from a module directly.
"""

Q8: What is the purpose of the with statement when handling files in Python?
"""
The with statement ensures that files are properly closed after their suite finishes, even if an exception occurs.
It is used for cleaner and safer file operations.
"""

Q9: What is the difference between multithreading and multiprocessing?
"""
Multithreading uses threads to run tasks concurrently within a single process, sharing memory.
Multiprocessing uses multiple processes to run tasks in parallel with separate memory space, better for CPU-bound tasks.
"""

"""

"""

Q11: What is memory management in Python?
"""
Memory management in Python involves allocating and deallocating memory to objects during execution.
Python manages memory using reference counting and garbage collection.
"""

Q12: What are the basic steps involved in exception handling in Python?
"""

Try block: Write the code that may raise an exception.

Except block: Handle the exception.

Else block (optional): Run if no exception occurs.

Finally block (optional): Cleanup code, runs no matter what.
"""

Q13: Why is memory management important in Python?
"""
Proper memory management prevents memory leaks, ensures program stability, and optimizes resource usage.
"""

Q14: What is the role of try and except in exception handling?
"""

try block wraps code that may raise an error.

except block catches and handles the error to prevent crashes.
"""

Q15: How does Python's garbage collection system work?
"""
Python uses reference counting and a cyclic garbage collector to detect and remove unused objects.
When an object's reference count reaches zero, it is automatically deleted.
"""

Q16: What is the purpose of the else block in exception handling?
"""
The else block runs if the try block does not raise any exception.
It is used to run code that should only execute when no errors occur.
"""

Q17: What are the common logging levels in Python?
"""

DEBUG: Detailed information for debugging.

INFO: Confirmation that things are working.

WARNING: Something unexpected happened.

ERROR: A more serious problem.

CRITICAL: Serious error that may prevent program continuation.
"""

Q18: What is the difference between os.fork() and multiprocessing in Python?
"""

os.fork(): Creates a child process (Unix only), low-level, requires manual management.

multiprocessing: High-level module, platform-independent, handles process creation and inter-process communication.
"""

Q19: What is the importance of closing a file in Python?
"""
Closing a file:

Frees system resources.

Ensures all data is written to disk.

Prevents data corruption or loss.
"""

Q20: What is the difference between file.read() and file.readline() in Python?
"""

file.read(): Reads the entire content of a file.

file.readline(): Reads a single line from the file at a time.
"""

Q21: What is the logging module in Python used for?
"""
The logging module is used to report status, error, and debugging messages during program execution.
It provides flexible logging levels and output formats.
"""

Q22: What is the os module in Python used for in file handling?
"""
The os module is used for interacting with the operating system:

File and directory manipulation

Path handling

Permissions and environment access
"""

Q23: What are the challenges associated with memory management in Python?
"""

Detecting and handling circular references.

Memory leaks due to lingering references.

Managing large datasets in memory efficiently.
"""

````python
# Assignment: File Handling, Logging, and Exception Handling in Python

# Q1: How can you open a file for writing in Python and write a string to it
with open("example1.txt", "w") as f:
    f.write("Hello, this is a test.")

# Q2: Write a Python program to read the contents of a file and print each line
with open("example1.txt", "r") as f:
    for line in f:
        print(line.strip())

# Q3: How would you handle a case where the file doesn't exist while trying to open it for reading
try:
    with open("nonexistent.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File does not exist.")

# Q4: Write a Python script that reads from one file and writes its content to another file
with open("source.txt", "w") as f:
    f.write("Copied content.")
with open("source.txt", "r") as src, open("destination.txt", "w") as dst:
    dst.write(src.read())

# Q5: How would you catch and handle division by zero error in Python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

# Q6: Write a Python program that logs an error message to a log file when a division by zero exception occurs
import logging
logging.basicConfig(filename="error_log.log", level=logging.ERROR)
try:
    a = 5 / 0
except ZeroDivisionError as e:
    logging.error("Division by zero error occurred: %s", e)

# Q7: How do you log information at different levels (INFO, ERROR, WARNING) in Python using the logging module
logging.basicConfig(filename="multi_level.log", level=logging.DEBUG)
logging.info("This is an info message.")
logging.warning("This is a warning message.")
logging.error("This is an error message.")

# Q8: Write a program to handle a file opening error using exception handling
try:
    with open("nofile.txt", "r") as f:
        print(f.read())
except IOError:
    print("Error: File could not be opened.")

# Q9: How can you read a file line by line and store its content in a list in Python
with open("example1.txt", "r") as f:
    lines = f.readlines()
print(lines)

# Q10: How can you append data to an existing file in Python
with open("example1.txt", "a") as f:
    f.write("\nThis is appended text.")

# Q11: Write a Python program that uses a try-except block to handle an error when attempting to access a dictionary key that doesn't exist
my_dict = {"a": 1}
try:
    print(my_dict["b"])
except KeyError:
    print("Key not found.")

# Q12: Write a program that demonstrates using multiple except blocks to handle different types of exceptions
try:
    x = int("abc")
    y = 10 / 0
except ValueError:
    print("Value error occurred.")
except ZeroDivisionError:
    print("Division by zero.")

# Q13: How would you check if a file exists before attempting to read it in Python
import os
if os.path.exists("example1.txt"):
    with open("example1.txt", "r") as f:
        print(f.read())
else:
    print("File does not exist.")

# Q14: Write a program that uses the logging module to log both informational and error messages
logging.basicConfig(filename="logfile.log", level=logging.DEBUG)
logging.info("Program started")
try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("Error occurred: %s", e)

# Q15: Write a Python program that prints the content of a file and handles the case when the file is empty
filename = "example1.txt"
with open(filename, "r") as f:
    content = f.read()
    if content:
        print(content)
    else:
        print("File is empty.")

# Q16: Demonstrate how to use memory profiling to check the memory usage of a small program
# Install memory_profiler in Colab: !pip install -q memory_profiler
from memory_profiler import profile

@profile
def memory_test():
    a = [i for i in range(10000)]
    return a

memory_test()

# Q17: Write a Python program to create and write a list of numbers to a file, one number per line
numbers = [1, 2, 3, 4, 5]
with open("numbers.txt", "w") as f:
    for num in numbers:
        f.write(f"{num}\n")

# Q18: How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging.handlers
rotating_logger = logging.getLogger("RotatingLog")
rotating_logger.setLevel(logging.INFO)
handler = logging.handlers.RotatingFileHandler("rotate.log", maxBytes=1_000_000, backupCount=3)
rotating_logger.addHandler(handler)
rotating_logger.info("Logging with rotation setup.")

# Q19: Write a program that handles both IndexError and KeyError using a try-except block
try:
    lst = [1, 2]
    print(lst[5])
    d = {"a": 1}
    print(d["b"])
except IndexError:
    print("Index out of range.")
except KeyError:
    print("Key not found in dictionary.")

# Q20: How would you open a file and read its contents using a context manager in Python
with open("example1.txt", "r") as f:
    print(f.read())

# Q21: Write a Python program that reads a file and prints the number of occurrences of a specific word
word = "test"
with open("example1.txt", "r") as f:
    text = f.read()
    count = text.count(word)
    print(f"The word '{word}' occurred {count} times.")

# Q22: How can you check if a file is empty before attempting to read its contents
if os.path.exists("example1.txt") and os.stat("example1.txt").st_size == 0:
    print("File is empty.")
else:
    with open("example1.txt", "r") as f:
        print(f.read())

# Q23: Write a Python program that writes to a log file when an error occurs during file handling
try:
    with open("maybe_missing.txt", "r") as f:
        data = f.read()
except FileNotFoundError as e:
    logging.error("Failed to open file: %s", e)
````
